INSTALADOR

Este código tem o objetivo de instalar as bibliotecas python e o modelo de LLM no Ollama.

Certifique-se que o arquivo config.toml esteja preenchido.

```python
import toml
import requests
```

```python
def url_llm():
    config = toml.load(r"config.toml")
    url = config['llm']['url']
    return url

# O nome do modelo é extraído do arquivo config.toml
def nome_modelo():
    config = toml.load(r"config.toml")
    nome_modelo = config['llm']['nome_modelo']
    return nome_modelo

def modelo_instalado(nome_modelo: str) -> bool:
    response = requests.get(rf"http://{url_llm()}:11434/api/tags")
    response.raise_for_status()
    
    modelos = response.json().get("models", [])
    nomes = [m["name"] for m in modelos]
    
    return nome_modelo in nomes

def instalar_modelo(nome_modelo: str):
    response = requests.post(
        rf"http://{url_llm()}:11434/api/pull",
        json={"name": nome_modelo}
    )
    response.raise_for_status()

def garantir_modelo(nome_modelo: str):
    if modelo_instalado(nome_modelo):
        print(f"✅ Modelo {nome_modelo} já está instalado.")
    else:
        print(f"⬇️ Instalando modelo {nome_modelo}...")
        instalar_modelo(nome_modelo)
        print("✅ Instalação concluída.")
```

```python
garantir_modelo(nome_modelo())
```
